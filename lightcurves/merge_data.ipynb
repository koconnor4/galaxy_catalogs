The lc photometric data tables I am going to merge with the sn host files at this  step. I do this here because the sn catalog table has the preliminary classification available (what I used to sort them by type) so that I know model to use. 

```python
# I want lc photometry data table if available to be included/merged...
# Previously had [ SN table, Nearby survey, [Host survey,elliptical radii away, other nearby radii away] ]
# Just going to append to the end of this  
# [ SN table, Nearby survey, [Host survey,elliptical radii away, other nearby radii away], [lc table] ]
# its appended as a list just in case there is more than one data table 

# need to keep track of all the lcs that get a catalog match so I know which ones I have that didn't get matched
# make sure they don't get left behind; I can still hold on to them and provide a fit
lc_match_idxs = []
# the catalog sn I'll know which don't have a lc match by virtue of the appended list will be empty (no data table there)

for i in range(len(candels_Ias)): 
    candels_Ias[i].append([]) # add it as a list incase there are multiple lc tables just append
    name = candels_Ias[i][0]['name'].lower()
    for j in range(len(lcs)):
        try:
            lc_name = lcs[j].meta['NAME'][0].lower()
        except:
            pass
            #print(j,'not NAME')
        try:
            lc_name = lcs[j].meta['IAUC'][0].lower()
        except:
            pass
            #print(j,'not IAUC,or NAME')
        try:
            lc_name = lcs[j].meta['SNID'][0].lower()
        except:
            pass
            #print(j,'Not SNID, or IAUC, or NAME???')
        finally: 
            if name == lc_name:
                print('we got a match!',i,name,j,lc_name)
                candels_Ias[i][-1].append(lcs[j])
                lc_match_idxs.append(j)

# Redumping... now have
# [sn table, nearby survey, [host survey,radii away, other nearby radii away], [lc table(s)] ]

pickle.dump(candels_Ias,open('data/merged/candels_Ias.pkl','wb'))
```
